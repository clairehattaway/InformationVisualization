```python
import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv('LEP_Populations.csv')
# Display the first few rows of the DataFrame
df.head()
```

```python
# Turn the 'LEP Population (Estimate)' column into a numeric type to filter languages with more than 0 speakers
df['LEP Population (Estimate)'] = df['LEP Population (Estimate)'].str.replace(',', '')
df['LEP Population (Estimate)'] = df['LEP Population (Estimate)'].astype(float)

# Remove languages with 0 speakers
df = df[df['LEP Population (Estimate)'] > 0]

# Turn the 'CVALEP Population (Estimate)' column into a numeric type
df['CVALEP Population (Estimate)'] = df['CVALEP Population (Estimate)'].str.replace(',', '')
df['CVALEP Population (Estimate)'] = df['CVALEP Population (Estimate)'].astype(float)
```

```python
# Now complete a few preliminary analyses to get an idea of the data before entering in Tableau for validation of the dashboard.
# % of LEP speakers who are also citizens of voting age (CVALEP)
per_cvalep = df['CVALEP Population (Estimate)'].sum() / df['LEP Population (Estimate)'].sum() * 100
print(f"Percentage of LEP speakers who are also citizens of voting age: {per_cvalep:.2f}%\n")

# Number of different languages spoken in NYC of LEP population
num_languages = df['Language'].nunique()
print(f"Number of different languages represented in NYC LEP Population: {num_languages}\n")

# Get total number of languages per borough
languages_per_borough = df.groupby('Borough')['Language'].nunique()
print(languages_per_borough, "\n")

# Get total number of langauges per community district
languages_per_community_district = df.groupby('Community District Name')['Language'].nunique()
print(languages_per_community_district.nlargest(5), '\n')
```

```python
# Get CVALEP % by language
per_cvalep = df.groupby('Language')['CVALEP Population (Estimate)'].sum() / df.groupby('Language')['LEP Population (Estimate)'].sum() * 100
print(per_cvalep.nlargest(5), '\n')

# Find top 5 langauges with the highest CVALEP % (percentage of LEP speakers who are also citizens of voting age) that are not 100%
per_cvalep_non_100 = per_cvalep[per_cvalep < 100]
top_5_non_100 = per_cvalep_non_100.nlargest(5)
print('Languages with highest CVALEP % (not 100%):', top_5_non_100)
```

```python
# Save the cleaned DataFrame to a new CSV file for use in Tableau
df.to_csv('LEP_Populations_Cleaned.csv', index=False)
```
